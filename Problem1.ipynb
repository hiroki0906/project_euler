```python
#お題:10未満の自然数のうち, 3 もしくは 5 の倍数になっているものは 3, 5, 6, 9 の4つがあり, これらの合計は 23 になる.
#同じようにして, 1000 未満の 3 か 5 の倍数になっている数字の合計を求めよ.
```

```python
import numpy as np
import time
```

```python
#list表記でやってみる
#手順(i)：①3の倍数リスト(list3)、②5の倍数リスト(list5)、論理表記で①or②のリストを作る
#手順(ii)：①or②のリストを合計する。
#※計算スピードも求め、numpyで計算してスピードを比較する。
```

```python
#list3を作成
start_list = time.time()
list3 = []
for i in range(1000):
    if i % 3 == 0:
        list3.append(i)
print(list3)
```

```python
#list5を作成
list5 = []
for i in range(1000):
    if i % 5 == 0:
        list5.append(i)
print(list5)
```

```python
#listの結合・重複削除・合計値算出
list3or5 = set(list3 + list5)
#一応自然数なので0も削除
list3or5.remove(0)
print(list3or5)
answer = sum(list3or5)
print(answer)
elapsed_time_list = time.time() - start_list
print("elapsed_time:{0}".format(elapsed_time_list))
```

```python
#numpyでやってみる
start_np = time.time()
number = np.array(range(1000))
print(number)
```

```python
list3or5 = number[np.logical_or(number % 3 == 0, number % 5 == 0)]
print(list3or5)
answer = np.sum(list3or5)
print(answer)
elapsed_time_np = time.time() - start_np
print("elapsed_time:{0}".format(elapsed_time_np))
```

```python
#関数で一般化してみる
def problem1(number, splitnumber1, splitnumber2):
    num = np.array(range(number))
    answer = np.sum(num[np.logical_or(num % splitnumber1 == 0, num % splitnumber2 == 0)])
    return answer
```

```python
result_prob1 = problem1(1000, 3, 5)
print(result_prob1)
```
